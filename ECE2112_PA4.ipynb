# <a id="Data-Wrangling"> Data Wrangling Problem </a>

```python
import pandas as pd

# Reads and stores the board2.xlsx to the variable board
board = pd.read_excel('board2.xlsx')
board
```

***
### <a id="a"> a. </a> Filename: Instru = [“Name”, “GEAS”, “Electronics >70”]; where track is constant as Instrumentation and hometown Luzon 

```python
# Using .loc, subsets the table board into 3 columns Name, GEAS, Electronics with the conditions of Hometown and Electronics
instru = board.loc[(board['Hometown']=='Luzon') & 
        (board['Electronics']>70),
        ['Name', 'GEAS', 'Electronics']]

instru
```

***
### <a id="b"> b </a> Filename: Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is constant as Mindanao and gender Female

```python
# Adds another column called 'Average" which gives the mean of the subjects
board['Average'] = board[['Math','Electronics','GEAS','Communication']].mean(axis=1)

# Using .loc, subsets the table 'board' into 4 columns, Name, Track, Electronics and Average with the set conditions
mindy = board.loc[(board['Hometown']=='Mindanao') & 
        (board['Gender']=='Female') & 
        (board['Average']>=55), 
        ['Name', 'Track', 'Electronics', 'Average']]

mindy
```

# <a id="Data-Visualization"> Data Visualization Problem </a>

## <a id="By-track"> By Track </a>

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
track = plt.bar(board['Track'], board['Average'])

# Set y-axis limits to zoom into the table
plt.ylim(65, 85)

plt.xlabel('Track')
plt.ylabel('Average')
plt.title('Average by Track')

# Show the plot
plt.show()
```

### Description: The table displays that students who pursued **Microelectronics** performed better in college, followed by **Communication** track and **Instrumantation** Track.
